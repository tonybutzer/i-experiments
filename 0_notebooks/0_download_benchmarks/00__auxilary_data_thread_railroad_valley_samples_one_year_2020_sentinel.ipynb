```python
import pandas as pd
def build_dancecard(items_s2):
    records = []
    for item in items_s2:
        key = item.id
        record = [key, 
                  item.assets["thumbnail"]["href"],
                  item.assets["metadata"]["href"],
                  item.assets["info"]["href"]
                 ]
        records.append(record)
        
    rdf = pd.DataFrame(records)
    return(rdf)
        
```

```python
kdf = build_dancecard(items_s2)
```

```python
def wget_to_s3(myid, s3_obj):
    print("wget this: %s",s3_obj)
    #file_obj_name = s3_obj.split('/')[-2] + '_' + s3_obj.split('/')[-1] # _ is for all files in one dir
    #file_obj_name = s3_obj.split('/')[-2] + '/' + s3_obj.split('/')[-1] # /is for subdur for each observation
    file_obj_name = s3_obj.split('/')[-1] 
    
    object_path = s3_obj.split('sentinel-s2-l2a-cogs')[-1]
    #obj_prefix = '/'.join(object_path.split('/')[:-2])
    #obj_name=obj_prefix + file_obj_name
    obj_name=file_obj_name
    dest_subdir = file_obj_name.split('.')
    dest_dir = 's3://dev-et-data/usmart/RAILROAD/sentinel-s2-l2a-cogs/' + myid +'/'
    cmd = 'wget -nv -O- ' + s3_obj + ' | aws s3 cp - ' + dest_dir 
    #print(cmd)
    #print(obj_name)
    cmd = cmd + obj_name
    print(cmd)
    result = subprocess_cmd(cmd)
    #result=0
    print(result)
```

```python
import math
def xfer_function_caller(row):
    start = time.time()
    logging.info("xfer Thread %s: starting", row)
    series = kdf.loc[row]
    myid = series[0]
    for i in range(1,len(series)):
        print('I is i is :\n',i)
        logging.info("sending %s", series[i])
        wget_to_s3(myid,series[i])
    
    #time.sleep(3)
    end = time.time()
    delta = end - start
    logging.info("Thread %s: finishing in %d seconds", row, math.trunc(delta))
```

```python
if __name__ == "__main__":
    format = "%(asctime)s: %(message)s"
    logging.basicConfig(format=format, level=logging.INFO,
                        datefmt="%H:%M:%S")

    #total_scenes = total_scenes - 120 # test fewer sometimes
    with concurrent.futures.ThreadPoolExecutor(max_workers=15) as executor:
        executor.map(xfer_function_caller, range(total_scenes))
        logging.info("Initial Threads %s: strted - i think i am waiting before any others", "executor")
```
