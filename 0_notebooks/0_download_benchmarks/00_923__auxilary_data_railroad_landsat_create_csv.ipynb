```python
def get_alternate_s3_dict(item, requested_assets):
    print(item.assets['thumbnail']['alternate']['s3']['href'])
```

```python
import pandas as pd
def build_dancecard(items_s2,requested_assets):
    print(requested_assets)
    records = []
    for item in items_s2:
        alternate_s3 = get_alternate_s3_dict(item, requested_assets)
        key = item.id
        record = [key,
                  item.properties['landsat:scene_id'],
                  item.properties['instruments'],
                  item.properties['landsat:cloud_cover_land'],
#                   item.assets["thumbnail"]["href"],
#                   item.assets["metadata"]["href"],
#                   item.assets["info"]["href"],
#                   item.assets["B04"]["href"],item.assets["B03"]["href"],item.assets["B02"]["href"],
#                   item.assets["B08"]["href"],
#                   item.assets["overview"]["href"],
                 ]
        asset_records = []
        for rasset in requested_assets:
            asset_records.append(item.assets[rasset]['alternate']['s3']['href'])
        for a in asset_records:
            record.append(a)
        records.append(record)
        
    rdf = pd.DataFrame(records)
    return(rdf)
        
```

```python
requested_assets =['thumbnail', 'reduced_resolution_browse', 'blue', 'green', 'red', 'nir08', 'ANG.txt', 'MTL.txt', 'MTL.xml', 'MTL.json', 'qa_pixel', 'qa_radsat']

kdf = build_dancecard(items_s2, requested_assets)
```

```python
kdf
```

```python
col_names = ['key', 'productID', 'instruments', 'coveragePct']

for a in requested_assets:
    col_names.append(a)
```

```python
kdf2 = kdf.set_axis(col_names, axis=1, inplace=False)
```

```python
kdf2.to_csv('./jon_landsat_all_of_2020_railroad_valley.csv')
```
