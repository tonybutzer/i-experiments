```python
import pandas as pd

csv_file = './jon_landsat_all_of_2020_railroad_valley.csv'
```

```python
df = pd.read_csv(csv_file)
df
```

```python
def get_basename(full_name):
    return(full_name.split('/')[-1])
```

```python
def copy_assets(myid, a_list):
    for a in a_list:
        print(get_basename(a))
        bucket_path = 's3://dev-et-data/usmart/RAILROAD/Landsat/'
        from_file = a
        to_file = bucket_path + myid + '/' + get_basename(a)
        cmd = 'aws s3 cp --request-payer requester ' + from_file + ' ' + to_file
        print(cmd)
        subprocess_cmd(cmd)
```

```python
for idx, row in df.iterrows():
    print(row[1])
    print(row[2])
    myid = row[1]
    copy_assets(myid, row[5:])
```
