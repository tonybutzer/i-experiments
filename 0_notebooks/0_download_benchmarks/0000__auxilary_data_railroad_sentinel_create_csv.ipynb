```python
import pandas as pd
def build_dancecard(items_s2):
    records = []
    for item in items_s2:
        key = item.id
        record = [key,
                  item.properties['sentinel:product_id'],
                  item.properties['sentinel:data_coverage'],
                  item.properties['eo:cloud_cover'],
                  item.assets["thumbnail"]["href"],
                  item.assets["metadata"]["href"],
                  item.assets["info"]["href"],
                  item.assets["B04"]["href"],item.assets["B03"]["href"],item.assets["B02"]["href"],
                  item.assets["B08"]["href"],
                  item.assets["overview"]["href"],
                 ]
        records.append(record)
        
    rdf = pd.DataFrame(records)
    return(rdf)
        
```

```python
kdf = build_dancecard(items_s2)
```

```python
kdf2 = kdf.set_axis(['key', 'productID','coveragePct', 'clouds','thumb', 'metadata', 'infoJSON','red','green','blue','nir','overview'], axis=1, inplace=False)
```

```python
kdf2.to_csv('./jon_sentinel_all_of_2020.csv')
```
